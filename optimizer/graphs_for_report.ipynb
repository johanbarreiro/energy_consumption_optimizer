```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the data
df = pd.read_csv("optimizer/evaluation/2024-07-10_21-31-14_evaluate_daily_output.csv")

# Extract the relevant columns
x = df['Date']
y1 = df['actual_energy_cost']
y2 = df['Optimized_Total_Cost']

# Calculate the total costs
total_actual_cost = y1.sum()
total_optimized_cost = y2.sum()

# Create a figure with two subplots side by side with different widths
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6), gridspec_kw={'width_ratios': [3, 1]})

# Line Chart
ax1.plot(x, y1, label='Actual Energy Cost')
ax1.plot(x, y2, label='Optimized Total Cost', linestyle='--')

ax1.set_xlabel('Date')
ax1.set_ylabel('Total Energy Cost (â‚¬) by day')
ax1.set_title('Actual Energy Cost vs Optimized Model Output Over Time')
ax1.legend()
ax1.tick_params(axis='x', rotation=45)

# Bar Graph
bars = ax2.bar(['Actual Energy Cost', 'Optimized Total Cost'], [total_actual_cost, total_optimized_cost], color=['blue', 'orange'])

ax2.set_ylabel('Total Energy Cost (â‚¬)')
ax2.set_title('Total Actual Energy Cost vs Total Optimized Model Output')

# Add data labels
for bar in bars:
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width() / 2.0, height, f'{height:.2f}', ha='center', va='bottom')

# Adjust layout
plt.tight_layout()
plt.show()
```
