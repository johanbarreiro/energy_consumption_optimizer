```python
import pandas as pd

# Load the dataset with semicolon as the delimiter
df = pd.read_csv('data/site_data/industrial_sites/industrial_site2.csv', delimiter=';')

# Extracting column names with 'Active Energy' and 'Active Power'
energy_columns = [col for col in df.columns if 'Active Energy' in col]
power_columns = [col for col in df.columns if 'Active Power' in col]

# Debug: Print the first few rows of the DataFrame
print("First few rows of the DataFrame:")
print(df.head())

# Debug: Print the columns found
print("Energy columns found:", energy_columns)
print("Power columns found:", power_columns)

# Initialize a results dictionary to store comparison results
comparison_results = {}
columns_order = []

# Iterate over energy columns and find the corresponding power columns
for energy_col in energy_columns:
    prefix = energy_col.replace('Active Energy', '')
    power_col = f"{prefix}Active Power"
    if power_col in df.columns:
        # Debug: Print the values before calculation
        print(f"Processing columns: {energy_col} and {power_col}")
        print("First few values of Active Energy column:", df[energy_col].head())
        print("First few values of Active Power column:", df[power_col].head())
        
        # Calculate the difference between 'Active Energy' and 4 times 'Active Power'
        comparison_results[energy_col] = df[energy_col] - 4 * df[power_col]
        
        # Debug: Print the calculated values
        print(f"First few values after calculation for {energy_col}:")
        print(comparison_results[energy_col].head())
        
        columns_order.extend([energy_col, power_col])
    else:
        # Debug: Print when a matching power column is not found
        print(f"Matching power column not found for: {energy_col}")

# Create a DataFrame from the comparison results
comparison_df = pd.DataFrame(comparison_results)

# Reorder the columns to ensure energy and power columns are next to each other
ordered_columns = []
for energy_col in energy_columns:
    prefix = energy_col.replace('Active Energy', '')
    power_col = f"{prefix}Active Power"
    if energy_col in comparison_df.columns and power_col in df.columns:
        ordered_columns.extend([energy_col, power_col])

comparison_df = comparison_df.reindex(columns=ordered_columns)

# Debug: Print the first few rows of the comparison DataFrame
print("First few rows of the comparison DataFrame:")
print(comparison_df.head())
```
